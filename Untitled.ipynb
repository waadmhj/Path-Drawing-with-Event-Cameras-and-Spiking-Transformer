# Viualize the Tracking of the Hand

```python
import cv2
import os

# Load the object coordinates from the text file
coordinates_file = 'logs/EVENT-Benchmark/result/EVENT/STNET/fullhand/fullhand_001.txt'

with open(coordinates_file, 'r') as f:
    object_coordinates = f.readlines()

# Load the video frames
video_path = 'data/output/val/fullhand'

# Iterate through the subfolders (neg and pos)
for subfolder in ['neg', 'pos']:
    subfolder_path = os.path.join(video_path, subfolder)

    # Get the frame files in the subfolder
    frame_files = sorted(os.listdir(subfolder_path))

    frame_id = 0
    for frame_file in frame_files:
        # Skip non-image files
        if not frame_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        frame_path = os.path.join(subfolder_path, frame_file)
        frame = cv2.imread(frame_path)

        # Check if the frame is None
        if frame is None:
            print(f"Failed to read frame: {frame_path}")
            continue

        # Get the object coordinates for the current frame
        if frame_id < len(object_coordinates):
            coords = object_coordinates[frame_id].strip().split(',')
            x, y, w, h = map(float, coords)

            # Draw bounding box on the frame
            cv2.rectangle(frame, (int(x), int(y)), (int(x + w), int(y + h)), (0, 255, 0), 2)

        print(f"Frame ID: {frame_id}, Frame shape: {frame.shape}")  # Debugging statement

        # Display the frame
        cv2.imshow('Bounding Box Visualization', frame)
        
        # Wait for a key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        frame_id += 1

# Release the window
cv2.destroyAllWindows()
```

# tracy and draw 

```python
import cv2
import numpy as np
import os

# Load the object coordinates from the text file
coordinates_file = 'logs/EVENT-Benchmark/result/EVENT/STNET/suv_hdr_1/suv_hdr_1_001.txt'

with open(coordinates_file, 'r') as f:
    object_coordinates = f.readlines()

# Load the video frames
video_path = 'data/img_120_split/val/suv_hdr_1'

# Initialize the previous center coordinates and the drawn image
prev_center = None
drawn_image = None

# Create a video writer
output_video_path = 'tracing_video.mp4'
frame_width, frame_height = None, None
output_video = None

# Iterate through the subfolders (neg and pos)
for subfolder in ['neg', 'pos']:
    subfolder_path = os.path.join(video_path, subfolder)

    # Get the frame files in the subfolder
    frame_files = sorted(os.listdir(subfolder_path))

    frame_id = 0
    for frame_file in frame_files:
        # Skip non-image files
        if not frame_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        frame_path = os.path.join(subfolder_path, frame_file)
        frame = cv2.imread(frame_path)

        # Check if the frame is None
        if frame is None:
            print(f"Failed to read frame: {frame_path}")
            continue

        # Get the object coordinates for the current frame
        if frame_id < len(object_coordinates):
            coords = object_coordinates[frame_id].strip().split(',')
            x, y, w, h = map(float, coords)

            # Calculate the center of the bounding box
            center_x = int(x + w / 2)
            center_y = int(y + h / 2)

            # Draw bounding box on the frame
            cv2.rectangle(frame, (int(x), int(y)), (int(x + w), int(y + h)), (0, 255, 0), 2)

            # Draw tracing path
            if prev_center is not None:
                cv2.line(frame, prev_center, (center_x, center_y), (0, 0, 255), 2)

            # Update the previous center coordinates
            prev_center = (center_x, center_y)

            # Draw on the image
            if drawn_image is None:
                drawn_image = np.zeros_like(frame)
            cv2.line(drawn_image, prev_center, (center_x, center_y), (0, 0, 255), 2)

        print(f"Frame ID: {frame_id}, Frame shape: {frame.shape}")  # Debugging statement

        # Display the frame
        cv2.imshow('Bounding Box Visualization', frame)

        # Write the frame to the video
        if output_video is None:
            frame_height, frame_width = frame.shape[:2]
            output_video = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'MJPG'), 30, (frame_width, frame_height))
        output_video.write(frame)

        # Wait for a key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        frame_id += 1

# Release the video writer
if output_video is not None:
    output_video.release()

# Display the final drawn image
cv2.imshow('Final Drawing', drawn_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Data Preprocessing 

## Red and blue

```python
import cv2
import numpy as np

# Specify the path to your video file
video_file = "regular/gt_events/C0182_evt.mp4"

# Create a VideoCapture object
cap = cv2.VideoCapture(video_file)

# Check if the video file is successfully opened
if not cap.isOpened():
    print("Error opening video file.")

# Get the frames per second (fps) and frame size
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Create blank images for positive (blue) and negative (red)
im_pos = np.zeros((frame_height, frame_width, 3), dtype=np.uint8)
im_neg = np.zeros((frame_height, frame_width, 3), dtype=np.uint8)

# Previous frame for calculating movement and brightness
prev_frame = None
prev_brightness = None

# Read and process each frame until the video ends
while cap.isOpened():
    # Read a frame from the video
    ret, frame = cap.read()

    # Check if the frame is successfully read
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Calculate the absolute difference between the current and previous frame
    if prev_frame is not None:
        frame_diff = cv2.absdiff(gray_frame, prev_frame)
    else:
        frame_diff = np.zeros_like(gray_frame)

    # Calculate the brightness difference between the current and previous frame
    if prev_brightness is not None:
        brightness_diff = gray_frame - prev_brightness
    else:
        brightness_diff = np.zeros_like(gray_frame)

    # Threshold the frame difference and brightness difference
    _, frame_diff_threshold = cv2.threshold(frame_diff, 30, 255, cv2.THRESH_BINARY)
    _, brightness_diff_threshold = cv2.threshold(brightness_diff, -30, 255, cv2.THRESH_BINARY_INV)

    # Combine the positive event positions from frame difference and brightness difference
    pos_indices = np.where((frame_diff_threshold > 0) | (brightness_diff_threshold > 0))

    # Set positive event positions in the positive image
    im_pos.fill(0)  # Reset positive image
    im_pos[pos_indices] = (255, 0, 0)  # Blue color (BGR)

    # Set negative event positions in the negative image
    im_neg.fill(0)  # Reset negative image
    neg_indices = np.where((frame_diff_threshold == 0) & (brightness_diff_threshold == 0))
    im_neg[neg_indices] = (0, 0, 255)  # Red color (BGR)

    # Display the frames
    cv2.imshow("Positive Events", im_pos)
    cv2.imshow("Negative Events", im_neg)

    # Update the previous frame and brightness
    prev_frame = gray_frame.copy()
    prev_brightness = gray_frame.copy()

    # Wait for the 'q' key to be pressed to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture object
cap.release()

# Close the windows
cv2.destroyAllWindows()
```

## generating frames

```python
import os
import cv2

# Path to the AVI video file
video_path = 'positive_events.mp4'

# Directory to save the extracted frames
output_dir = 'pos'

# Open the video file
video = cv2.VideoCapture(video_path)

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Initialize frame count
frame_count = 0

# Read and save frames until the video ends
while video.isOpened():
    # Read the next frame
    ret, frame = video.read()

    # Break if no frame is retrieved
    if not ret:
        break

    # Save the frame as an image file
    frame_filename = "{:05d}.jpg".format(frame_count + 1)
    frame_path = os.path.join(output_dir, frame_filename)
    cv2.imwrite(frame_path, frame)

    # Increment frame count
    frame_count += 1

# Release the video capture object and close the output directory
video.release()
cv2.destroyAllWindows()
```
